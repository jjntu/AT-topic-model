```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import json
```

```python
hyper_min = 10
hyper_max = 290

data_list = {
    'hyper': [i for i in range(hyper_min, hyper_max + 10, 10)],
    'coherence_min': [],
    'diversity_min': [],
    'topic_min': [],
    'coherence_nr': [],
    'diversity_nr': [],
    'topic_nr': []
}

min_path = [f"result/10-None_15-5-0.0-{_min}_2-0.01.0_1-1.json" for _min in range(hyper_min, hyper_max + 10, 10)]

for path in min_path:
    with open(path, 'r') as file:
        data = json.load(file)
        data_list["coherence_min"].append(data[0]["Scores"]["c_v"])
        data_list["diversity_min"].append(data[0]["Scores"]["diversity"])
        data_list["topic_min"].append(data[0]["Topic Number"])

nr_path = [f"result/10-{nr}_15-5-0.0-20_2-0.01.0_1-2.json" for nr in range(hyper_min, hyper_max + 10, 10)]

for path in nr_path:
    with open(path, 'r') as file:
        data = json.load(file)
        data_list["coherence_nr"].append(data[0]["Scores"]["c_v"])
        data_list["diversity_nr"].append(data[0]["Scores"]["diversity"])
        data_list["topic_nr"].append(data[0]["Topic Number"])
        
data_list
```

```python
#  data_list
data = data_list

# 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

#  coherence  diversity 
ax1.plot(data['hyper'], data['coherence_min'], label='coherence_min')
ax1.plot(data['hyper'], data['diversity_min'], label='diversity_min')
ax1.plot(data['hyper'], data['coherence_nr'], label='coherence_nr')
ax1.plot(data['hyper'], data['diversity_nr'], label='diversity_nr')

# 
ax1.set_xlabel('min_cluster_size, nr_topics')
ax1.set_ylabel('coherence, diversity')
ax1.set_xlim(0, 300)
ax1.legend()

#  topic number 
ax2.plot(data['hyper'], data['topic_min'], label='topic number_min')
ax2.plot(data['hyper'], data['topic_nr'], label='topic number_nr')

# 
ax2.set_xlabel('min_cluster_size, nr_topics')
ax2.set_ylabel('topic number')
ax2.set_xlim(0, 300)
ax2.legend()

# 
plt.tight_layout()
plt.show()
```
